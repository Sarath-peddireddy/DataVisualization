```python
import pandas as pd
data = pd.read_csv('student.csv')
data
```

```python
data.isnull().sum()
```

```python
print(data.shape)
print(data.info())
```

#                                                               Visualize Data 

## Average GPA by Ethnicity

```python
average_gpa_by_ethnicity = data.groupby('Ethnicity')['GPA'].mean()
print("Average GPA by Ethnicity:")
print(average_gpa_by_ethnicity)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.barplot(x=average_gpa_by_ethnicity.index, y=average_gpa_by_ethnicity.values)
plt.title('Average GPA by Ethnicity')
plt.xlabel('Ethnicity')
plt.ylabel('Average GPA')
plt.xticks(rotation=45)
plt.show()
```

## Distribution of GPA

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.histplot(data['GPA'], bins=20, kde=True)
plt.title('Distribution of GPA')
plt.xlabel('GPA')
plt.ylabel('Frequency')
plt.show()
```

## Box Plot of GPA by Gender

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Gender', y='GPA', data=data)
plt.title('GPA Distribution by Gender')
plt.xlabel('Gender')
plt.ylabel('GPA')
plt.show()
```

## Bar Plot of Average GPA by Parental Education

```python
plt.figure(figsize=(10, 6))
average_gpa_by_parental_education = data.groupby('ParentalEducation')['GPA'].mean()
sns.barplot(x=average_gpa_by_parental_education.index, y=average_gpa_by_parental_education.values)
plt.title('Average GPA by Parental Education')
plt.xlabel('Parental Education')
plt.ylabel('Average GPA')
plt.xticks(rotation=45)
plt.show()
```

## Count Plot of Extracurricular Participation

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='Extracurricular', data=data, palette='viridis')
plt.title('Count of Extracurricular Participation')
plt.xlabel('Participates in Extracurricular Activities')
plt.ylabel('Number of Students')
plt.show()
```

## Scatter Plot of Study Time vs. GPA

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='StudyTimeWeekly', y='GPA', data=data, hue='Gender', alpha=0.7)
plt.title('Study Time vs. GPA')
plt.xlabel('Study Time (Hours per Week)')
plt.ylabel('GPA')
plt.legend(title='Gender')
plt.show()
```

##  Stacked Bar Chart of Sports and Music Participation

```python
sports_music_counts = data.groupby(['Sports', 'Music']).size().unstack()
sports_music_counts.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title('Participation in Sports and Music')
plt.xlabel('Sports Participation')
plt.ylabel('Number of Students')
plt.xticks(rotation=0)
plt.legend(title='Music Participation')
plt.show()
```

## Heatmap of Correlation Matrix

```python
plt.figure(figsize=(12, 8))
correlation_matrix = data.corr()
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm', square=True)
plt.title('Correlation Heatmap')
plt.show()
```

## Pie Chart of Extracurricular Participation

```python
plt.figure(figsize=(4, 4))
extracurricular_counts = data['Extracurricular'].value_counts()
plt.pie(extracurricular_counts, labels=extracurricular_counts.index, autopct='%1.1f%%', startangle=90, colors=sns.color_palette('pastel'))
plt.title('Extracurricular Participation')
plt.axis('equal')  # Equal aspect ratio ensures that pie chart is circular
plt.show()
```

## Pair Plot of Key Features

```python
sns.pairplot(data[['GPA', 'Absences', 'StudyTimeWeekly', 'Gender']], diag_kind='kde', hue='Gender', palette='Set1')
plt.title('Pair Plot of Key Features')
plt.show()
```
